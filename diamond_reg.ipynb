```python
import pandas as pd
```

```python
# To perform machine learning, we need at least 1 python package and 2 of its modules
from sklearn.linear_model import LinearRegression # linear regression

# Train-test split package
from sklearn.model_selection import train_test_split # split your dataset
```

```python
data_frame = pd.read_csv("diamonds_regression.csv")
```

```python
data_frame.head(5)
```

```python
columns = ['carat', 'cut', 'color', 'clarity', 'depth', 'table']
```

```python
target = 'price'
```

```python
X = data_frame[columns]
y = data_frame[target]
```

```python
X
```

```python
X_dummies = pd.get_dummies(X)
```

```python
X_dummies
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_dummies, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
filename = 'diamonds_regression.sav'
```

```python
import joblib
joblib.dump(model, filename)
```
